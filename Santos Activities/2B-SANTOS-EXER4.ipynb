# **Import Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
```

# **1. Data Exploration:**

```python
data_path = "Lesson_4_student_grades_dataset.csv"
dataset = pd.read_csv(data_path)
```

# **2. Data Preprocessing:**

```python
print(dataset)

X = dataset.drop('Grades', axis=1)
Y = dataset[['Grades']]
```

# **3. Train-Test Split:**

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

model = LinearRegression()
```

# **4. Linear Regression Model:**

```python
a = dataset[['Hours_of_Study']]
b = dataset[['Attendance_Percentage']]
c = dataset[['Previous_Exam_Scores']]
d = dataset[['Extracurricular_Activities']]
e = dataset[['Travel_Time_to_School']]
Y = dataset[['Grades']]

model1 = LinearRegression()
model2 = LinearRegression()
model3 = LinearRegression()
model4 = LinearRegression()
model5 = LinearRegression()

model1.fit(a, Y)
model2.fit(b, Y)
model3.fit(c, Y)
model4.fit(d, Y)
model5.fit(e, Y)
```

```python
plt.scatter(a, Y)
plt.plot(a, model1.predict(a), color = 'red')
plt.xlabel('Independent Variable (X)')
plt.ylabel('Dependent Variable (Y)')
plt.title("Linear Regression")
plt.show()

plt.scatter(b, Y)
plt.plot(b, model2.predict(b), color = 'red')
plt.xlabel('Independent Variable (X)')
plt.ylabel('Dependent Variable (Y)')
plt.title("Linear Regression")
plt.show()

plt.scatter(c, Y)
plt.plot(c, model3.predict(c), color = 'red')
plt.xlabel('Independent Variable (X)')
plt.ylabel('Dependent Variable (Y)')
plt.title("Linear Regression")
plt.show()

plt.scatter(d, Y)
plt.plot(d, model4.predict(d), color = 'red')
plt.xlabel('Independent Variable (X)')
plt.ylabel('Dependent Variable (Y)')
plt.title("Linear Regression")
plt.show()

plt.scatter(e, Y)
plt.plot(e, model5.predict(e), color = 'red')
plt.xlabel('Independent Variable (X)')
plt.ylabel('Dependent Variable (Y)')
plt.title("Linear Regression")
plt.show()
```

# **5. Model Training:**

```python
model.fit(X_train, Y_train)
```

# **6. Predictions and Evaluation:**

```python
new_data = np.array([[2, 70, 85, 1, 50]])

prediction = model.predict(new_data)

print(f"Predict Grade: {prediction[0]}")
```

# **7. Feature Importance Analysis:**

```python
print("Hours of Study")
slope = model1.coef_[0]
intercept = model1.intercept_
print('Slope', slope)
print('Intercept', intercept)
print(" ")
print("Attendance Percentage")
slop = model2.coef_[0]
intercep = model2.intercept_
print('Slope', slop)
print('Intercept', intercep)
print(" ")
print("Previous Exam Scores")
slo = model3.coef_[0]
interce = model3.intercept_
print('Slope', slo)
print('Intercept', interce)
print(" ")
print("ExtraCurricular Activities")
sl = model4.coef_[0]
interc = model4.intercept_
print('Slope', sl)
print('Intercept', interc)
print(" ")
print("Travel Time to School")
s = model5.coef_[0]
inter = model5.intercept_
print('Slope', s)
print('Intercept', inter)
```
